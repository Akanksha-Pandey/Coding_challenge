```python
# combine metadata and cluster table based on Sample column
combined = pd.merge(left = metadata, right = cluster_table, on = 'Sample')
```

```python
combined.head()
```

```python
#Create two different dataframes based on timing of FMT treatment
before_FMT = combined.loc[combined['Time'] == 0,:]
before_FMT = before_FMT.drop(['Time','Sample'], axis=1)
before_FMT = before_FMT.set_index('Patient')
```

```python
after_FMT = combined.loc[combined['Time'] == 1,:]
after_FMT = after_FMT.drop(['Time','Sample'], axis=1)
after_FMT = after_FMT.set_index('Patient')
```

```python
#Calculate the differences in read counts for different clusters in before and after FMT samples
#difference = before_FMT - after_FMT
```
